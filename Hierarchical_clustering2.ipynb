```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from sklearn import manifold,datasets
```

```python
x1,y1=make_blobs(n_samples=300,centers=[[2,5],[1,6],[4,2],[2,8]],cluster_std=0.9)
```

```python
plt.scatter(x1[:,0],x1[:,1],marker="o")
```

```python
agglom=AgglomerativeClustering(n_clusters=4,linkage="average")
```

```python
agglom.fit(x1,y1)
```

```python
plt.figure(figsize=(8,6))

x_min,x_max=np.min(x1,axis=0),np.max(x1,axis=0)#create maximum and minimum range of x1

x1=(x1-x_min)/(x_max-x_min)# Get the average distance for X1.

for i in range(x1.shape[0]):
  plt.text(x1[i,0],x1[i,1],str(y1[i]),color=plt.cm.nipy_spectral(agglom.labels_[i]/10.),fontdict={'weight':'bold',"size":9})


plt.xticks([])
plt.yticks([])

plt.scatter(x1[:,0],x1[:,1],marker=".")
plt.show()

```

```python
dist_matrix=distance_matrix(x1,x1)
print(dist_matrix)
```

```python
z=hierarchy.linkage(dist_matrix,'complete')
```

```python
dendro=hierarchy.dendrogram(z)
```
